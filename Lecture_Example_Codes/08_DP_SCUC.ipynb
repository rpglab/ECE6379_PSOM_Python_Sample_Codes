```python
from pyomo.environ import *

# Create model
model = ConcreteModel()

# Parameters
C1 = 10
SU1 = 800
Pgmin1 = 30
Pgmax1 = 80

C2 = 30
SU2 = 100
Pgmin2 = 50
Pgmax2 = 90

# Declare Variables - Generator 1
model.u11 = Var(within=Binary)
model.v11 = Var(within=Binary)
model.P11 = Var()
model.u12 = Var(within=Binary)
model.v12 = Var(within=Binary)
model.P12 = Var()

# Declare Variables - Generator 2
model.u21 = Var(within=Binary)
model.v21 = Var(within=Binary)
model.P21 = Var()
model.u22 = Var(within=Binary)
model.v22 = Var(within=Binary)
model.P22 = Var()

# Define objective function
model.obj = Objective(
    expr=(SU1*model.v11 + C1*model.P11 + SU1*model.v12 + C1*model.P12) + 
         (SU2*model.v21 + C2*model.P21 + SU2*model.v22 + C2*model.P22),
    sense=minimize
)

# Generator 1 constraints
model.eq11 = Constraint(expr=Pgmin1*model.u11 <= model.P11)
model.eq12 = Constraint(expr=Pgmin1*model.u12 <= model.P12)
model.eq13 = Constraint(expr=model.P11 <= Pgmax1*model.u11)
model.eq14 = Constraint(expr=model.P12 <= Pgmax1*model.u12)
model.eq15 = Constraint(expr=model.v11 >= model.u11 - 0)
model.eq16 = Constraint(expr=model.v12 >= model.u12 - model.u11)

# Generator 2 constraints
model.eq21 = Constraint(expr=Pgmin2*model.u21 <= model.P21)
model.eq22 = Constraint(expr=Pgmin2*model.u22 <= model.P22)
model.eq23 = Constraint(expr=model.P21 <= Pgmax2*model.u21)
model.eq24 = Constraint(expr=model.P22 <= Pgmax2*model.u22)
model.eq25 = Constraint(expr=model.v21 >= model.u21 - 0)
model.eq26 = Constraint(expr=model.v22 >= model.u22 - model.u21)

# Power balance constraints
model.PowerBalance1 = Constraint(expr=model.P11 + model.P21 == 90)
model.PowerBalance2 = Constraint(expr=model.P12 + model.P22 == 70)

# Solver setting
solver = SolverFactory('gurobi')
solver.options['mipgap'] = 0.0
solver.options['timelimit'] = 90

# Solve
results = solver.solve(model, tee=True)

# Display results
print(f"u11 = {model.u11.value}, P11 = {model.P11.value}")
print(f"u12 = {model.u12.value}, P12 = {model.P12.value}")
print()
print(f"u21 = {model.u21.value}, P21 = {model.P21.value}")
print(f"u22 = {model.u22.value}, P22 = {model.P22.value}")
```
