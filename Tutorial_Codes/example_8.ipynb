```python
from pyomo.environ import *

# Create model
model = ConcreteModel()

# Declare Variables
model.x1 = Var()
model.x2 = Var()

# Define objective function
model.obj = Objective(expr=0, sense=maximize)

# Define constraints
model.constName = Constraint(expr=3*model.x1 + 2*model.x2 == 9)
model.constName_2 = Constraint(expr=model.x1 + 6*model.x2 == -7)

# Solver setting
solver = SolverFactory('gurobi')
solver.options['mipgap'] = 0.0
solver.options['timelimit'] = 90

# Solve
results = solver.solve(model, tee=True)

# Display results
print(f"x1 = {model.x1.value}")
print(f"x2 = {model.x2.value}")
```
