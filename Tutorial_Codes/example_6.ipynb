```python
from pyomo.environ import *

# Create model
model = ConcreteModel()

# Declare Variables
model.x1 = Var(within=NonNegativeReals)
model.x2 = Var(within=NonNegativeReals)
model.x3 = Var(within=NonNegativeReals)

# Define objective function
model.obj = Objective(expr=2*model.x1 + 5*model.x2 + model.x3, sense=maximize)

# Define constraints
model.constName = Constraint(expr=model.x1 + 3*model.x2 + 2*model.x3 <= 10)
model.constName_2 = Constraint(expr=2*model.x1 + model.x2 + 5*model.x3 <= 8)

# Solver setting
solver = SolverFactory('gurobi')
solver.options['mipgap'] = 0.0
solver.options['timelimit'] = 90

# Solve
results = solver.solve(model, tee=True)

# Show me the results
print(f"x1 = {model.x1.value}")
print(f"x2 = {model.x2.value}")
print(f"x3 = {model.x3.value}")
```
